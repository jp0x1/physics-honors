# Dynamics

## Introduction / History

Prior to Newton's Principia Mathematica, people knew that objects move throughout space with a position, velocity, and acceleration. However, people were unable to understand what caused these objects to move. 

Newton discovered that we can find the origin of what causes objects to move throughout space through a term called "Force(s)".

"Force" basically is the "push/pull" that moves the object throughout space. You may know these "things" as:

- Your hand pushing on a block, causing it to move: Your hand acts as the force moving the object

- You pulling on a sled, which causes the sled to move: The sled acts as the force moving the object

- You throwing a ball, which causes the ball to move: Your throw acts as the force moving the object

When Newton published the Principia Mathematica, he defined three fundamental laws of motion often know as "Newton's Laws"

The laws are as defined:

1. Objects in motion will stay in motion, unless acted upon by an external force. 

2.  $\sum{F} = ma$

3. For every reaction, there is an opposite and equal reaction


These three are basically the backbone of all dynamics, as well as physics

#### 1. Objects in motion will stay at rest or stay in motion, unless acted upon by an external force. 

Essentially, this law defines that objects will either not move, or move at a constant velocity until you apply an external force. 

We can see this in everyday life:

- A concrete block won't move until you push it, applying an external force to the block and moving it 

- An air hockey puck will continue to slide at a near constant velocity after you shoot it because there is little to no external force acting on the puck causing it to move faster or causing it to stop 

Mathematically, we can see this through kinematics. 

If we think of force as the origin of acceleration, as shown
$$F = ma $$
$$a = F/m$$

and think back towards our velocity kinematic equation 
$$v = {v}_0 - a(t)$$

If we have no external force, or $F=0$, we therefore cannot have an acceleration. 
$$ a = 0/m $$
$$v = {v}_0 - 0(t)$$
$$v = {v}_0$$

This proves mathematically that without a force, objects either move at constant speed (no change from initial to final velocity) or not move at all (initial velocity starts at zero)

#### 2.  $\sum{F} = ma$

This equation represents that:

$$ \sum{Force}({N}) = Mass({kg}) * Acceleration ({m/s^2}) $$

We can calculate force as the mass multiplied by the acceleration, and can rearrange the equation to find the other variables for other scenarios.

If your problem involves this, simply plug in values for mass, force, or acceleration and solve for the unknown quantity

An important thing to consider is that this represents the NET FORCE, or the **sum of forces**. 

* It is important to remember that force is a VECTOR, so forces can be in opposite directions of each other. What we can do is simply subtract one force from another if they are in opposite directions. Same goes for forces in the same direction

Example 1: 

Force ${F}_a$ pulls the object to the left, ${F}_b$ pulls the object to the right, we simply ${F}_a$ - ${F}_b$ in order to find the net force of the object

Visually we can use a tug of war as an example. Imagine you and your friend are pulling from both sides of a rope. You both are exerting force in opposite directions, and the rope begins moving towards you. You know that you are pulling with all of your might, yet the rope moves slowly towards you. This is because the NET FORCE \sum{Force}({N}) is represented as your force (moving towards you) - friend's force (moving away), producing a smaller net force, causing the rope to move slower

Example 2:

Force ${F}_a$ pushes the object to the left, ${F}_b$ pushes the object to the left as well, we simply ${F}_a$ + ${F}_b$ in order to find the net force of the object

You and your friend are trying to move a box. You and your friend push, you realize that the object is moving faster than if you had just pushed it yourself. This is because the net force is your force + friend's force, producing a larger net force moving the object faster

##### Weight

Weight essentially is the force a planet pulls down on you. 

**IMPORTANT** - It is important that you do not confuse weight with mass, weight is the force of the planet pulling on you, and mass is a number that represents a measurement of your inertia. Your mass throughout the universe will be the same, while your weight can change depending on the force of gravity pulling you down.

We define "weight" as $$F = m * g$$

F represents the **force of gravity or weight** (N),

m represents the mass (g),

g represents the **acceleration due to gravity** ($m/s^2$)

**The Earth's acceleration due to gravity will always be $9.8m/s^2$**

This proves useful when you need to find either weight or mass of an object (ON EARTH). (Simply input whatever is provided with the acceleration due to gravity)

Problems that involve weight will often ask for the weight on different planets (WILL EXPLICITLY BE MENTIONED, IF NO PLANET IS NAMED, ASSUME ITS EARTH).

What we simply need to do is input the acceleration due to gravity for different planets (will ALWAYS be provided)
